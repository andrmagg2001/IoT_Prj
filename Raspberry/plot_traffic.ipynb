# Plot traffic

```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

CSV_DATA = pd.read_csv("../logs/day4.csv")
CSV_DATA["ts"] = pd.to_datetime(CSV_DATA["ts"])
CSV_DATA["cmd_code"] = CSV_DATA["cmd"].apply(lambda x: 0 if "G1" in x else 1)  
```

```python

plt.figure(figsize=(10,4))
plt.plot(CSV_DATA["ts"], CSV_DATA["cmd_code"], drawstyle="steps-post", label="Command (0=G1, 1=G2)")
plt.yticks([0,1], ["G1 (TL1 verde)", "G2 (TL2 verde)"])
plt.xlabel("Time")
plt.ylabel("Active Signal")
plt.title("Traffic Light Switching Over Time")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
